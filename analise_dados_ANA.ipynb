A - A palavra foi escrita ortograficamente. Deve-se considerar também a escrita de uma palavra do mesmo campo semântico. No caso de frase ou expressão, corrigir apenas a palavra em foco

B - A grafia da palavra apresenta desvios ortográficos: transcrição da fala, escrita com substituição de
letras, inversão na ordem das letras ou omissão de letras, sem comprometer a leitura da palavra

C - A escrita apresenta uma letra para cada sílaba ou oscila entre uma letra representando uma sílaba e
sílabas representadas por mais de uma letra. Deve-se garantir que haja relação entre a letra grafada
e a pauta sonora da sílaba

BR - Folha totalmente em branco

NL - Nulo (Palavra nada relacionada à que foi dada a escrever (fora do campo semântico); escrita incompreensível: letras sem relação com a pauta sonora da sílaba; desenhos ou rabiscos na tentativa de imitar a escrita.)
